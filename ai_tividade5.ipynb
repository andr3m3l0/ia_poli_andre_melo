```python
import numpy as np
import matplotlib.pyplot as plt

# Função de ativação sigmoide e sua derivada
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_deriv(x):
    return x * (1 - x)

# Definição das portas lógicas
logic_gates = {
    'AND': {
        'X': np.array([[0, 0], [0, 1], [1, 0], [1, 1]]),
        'y': np.array([[0], [0], [0], [1]])
    },
    'OR': {
        'X': np.array([[0, 0], [0, 1], [1, 0], [1, 1]]),
        'y': np.array([[0], [1], [1], [1]])
    },
    'XOR': {
        'X': np.array([[0, 0], [0, 1], [1, 0], [1, 1]]),
        'y': np.array([[0], [1], [1], [0]])
    }
}

# Hiperparâmetros para testar
hidden_neurons_options = [4, 8, 12]
epoch_options = [1000, 5000, 10000]
tolerance_options = [0.1, 0.01, 0.001]

# Resultados
resultados = []

# Loop para cada porta lógica
for gate_name, data in logic_gates.items():
    X = data['X']
    y = data['y']
    
    for hidden_size in hidden_neurons_options:
        for max_epochs in epoch_options:
            for tol in tolerance_options:

                print(f"\nTreinando {gate_name} | Neurônios ocultos: {hidden_size} | Épocas: {max_epochs} | Tolerância: {tol}")
                
                # Inicialização dos pesos e bias
                np.random.seed(42)
                input_size = 2
                output_size = 1

                W1 = np.random.uniform(-1, 1, (input_size, hidden_size))
                b1 = np.random.uniform(-1, 1, (1, hidden_size))

                W2 = np.random.uniform(-1, 1, (hidden_size, output_size))
                b2 = np.random.uniform(-1, 1, (1, output_size))

                lr = 0.1
                loss_history = []

                convergiu = False

                for epoch in range(max_epochs):
                    # FORWARD
                    z1 = np.dot(X, W1) + b1
                    a1 = sigmoid(z1)
                    z2 = np.dot(a1, W2) + b2
                    a2 = sigmoid(z2)

                    # ERRO
                    error = y - a2
                    loss = np.mean(error**2)
                    loss_history.append(loss)

                    if loss <= tol:
                        convergiu = True
                        break

                    # BACKPROP
                    delta2 = error * sigmoid_deriv(a2)
                    dW2 = np.dot(a1.T, delta2)
                    db2 = np.sum(delta2, axis=0, keepdims=True)

                    delta1 = np.dot(delta2, W2.T) * sigmoid_deriv(a1)
                    dW1 = np.dot(X.T, delta1)
                    db1 = np.sum(delta1, axis=0, keepdims=True)

                    # ATUALIZAÇÃO
                    W2 += lr * dW2
                    b2 += lr * db2
                    W1 += lr * dW1
                    b1 += lr * db1

                # Armazenar resultados
                final_preds = sigmoid(np.dot(sigmoid(np.dot(X, W1) + b1), W2) + b2)
                accuracy = np.mean(np.round(final_preds) == y)

                resultados.append({
                    'porta': gate_name,
                    'neurônios_ocultos': hidden_size,
                    'épocas': epoch + 1,
                    'tolerância': tol,
                    'convergiu': convergiu,
                    'loss_final': loss,
                    'acurácia': accuracy
                })

                # Mostrar predições finais
                print(f"Convergiu: {convergiu} | Loss final: {loss:.4f} | Acurácia: {accuracy:.2f}")
                
                # Exibir gráfico da curva de erro
                plt.figure(figsize=(6, 4))
                plt.plot(loss_history)
                plt.title(f'{gate_name} | Ocultos: {hidden_size} | Épocas: {epoch + 1}')
                plt.xlabel('Época')
                plt.ylabel('Loss (MSE)')
                plt.grid(True)
                plt.tight_layout()
                plt.show()

# Exibir tabela resumo (pode exportar para pandas ou salvar em CSV se desejar)
import pandas as pd

df_resultados = pd.DataFrame(resultados)
print("\nResumo dos testes:")
print(df_resultados)
```

```python
!pip install -q tensorflow

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Dados das portas lógicas
logic_gates = {
    'AND': {
        'X': np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32),
        'y': np.array([[0], [0], [0], [1]], dtype=np.float32)
    },
    'OR': {
        'X': np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32),
        'y': np.array([[0], [1], [1], [1]], dtype=np.float32)
    },
    'XOR': {
        'X': np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32),
        'y': np.array([[0], [1], [1], [0]], dtype=np.float32)
    }
}

# Hiperparâmetros para testar
hidden_neurons_options = [4, 8, 12]
epoch_options = [1000, 5000, 10000]
tolerance_options = [0.1, 0.01, 0.001]

# Resultados
resultados_tf = []

# Loop pelos testes
for gate_name, data in logic_gates.items():
    X = data['X']
    y = data['y']

    for hidden_size in hidden_neurons_options:
        for max_epochs in epoch_options:
            for tol in tolerance_options:

                # Modelo sequencial
                model = Sequential([
                    Dense(hidden_size, input_dim=2, activation='sigmoid'),
                    Dense(1, activation='sigmoid')
                ])

                # Compilação do modelo com otimizador SGD
                model.compile(optimizer=SGD(learning_rate=0.9), loss='mse')

                # Callback para parada antecipada com base no loss
                early_stop = tf.keras.callbacks.EarlyStopping(
                    monitor='loss',
                    patience=200,
                    min_delta=tol,
                    restore_best_weights=True,
                    verbose=0
                )

                # Treinamento
                history = model.fit(X, y, epochs=max_epochs, verbose=0, callbacks=[early_stop])

                final_loss = history.history['loss'][-1]
                preds = model.predict(X)
                acc = np.mean(np.round(preds) == y)

                convergiu = len(history.history['loss']) < max_epochs

                print(f"{gate_name} | Neurônios: {hidden_size} | Épocas: {len(history.history['loss'])} | Tolerância: {tol}")
                print(f"Convergiu: {convergiu} | Loss final: {final_loss:.4f} | Acurácia: {acc:.2f}")

                resultados_tf.append({
                    'porta': gate_name,
                    'neurônios_ocultos': hidden_size,
                    'épocas': len(history.history['loss']),
                    'tolerância': tol,
                    'convergiu': convergiu,
                    'loss_final': final_loss,
                    'acurácia': acc
                })

                # Gráfico da curva de erro
                plt.figure(figsize=(6, 4))
                plt.plot(history.history['loss'])
                plt.title(f'{gate_name} | Ocultos: {hidden_size} | Épocas: {len(history.history["loss"])}')
                plt.xlabel('Época')
                plt.ylabel('Loss (MSE)')
                plt.grid(True)
                plt.tight_layout()
                plt.show()

# Exibir tabela resumo dos testes
df_tf = pd.DataFrame(resultados_tf)
print("\nResumo dos testes com TensorFlow:")
print(df_tf)
```
